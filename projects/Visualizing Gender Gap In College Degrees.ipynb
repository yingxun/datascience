```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('C:/Users/User/Datasets/percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Engineering', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005,87,"Men")
        ax.text(2002,8,"Women")
    elif sp==5:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
plt.show()
```

```python
#include a total of 17 line charts and display each category in an individual column with descending percentages of degrees 
#awarded to women

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#change figsize to accommodate a 6x3 grid
fig = plt.figure(figsize=(18, 18))
#include counter for each list
a=0
b=0
c=0

#add_subplot index range from 1 onwards so we set the range to be 1-18
for sp in range(1,19):
        if (sp in (1,4,7,10,13,16)):
                ax = fig.add_subplot(6,3,sp)
                ax.plot(women_degrees['Year'], women_degrees[stem_cats[a]], c=cb_dark_blue, label='Women', linewidth=3)
                ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[a]], c=cb_orange, label='Men', linewidth=3)
                ax.set_title(stem_cats[a])
                #increment list counter by 1
                a+=1

        elif sp==17:
            continue
        elif sp in (2,5,8,11,14):
                ax = fig.add_subplot(6,3,sp)
                ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[b]], c=cb_dark_blue, label='Women', linewidth=3)
                ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[b]], c=cb_orange, label='Men', linewidth=3)
                ax.set_title(lib_arts_cats[b])
                #increment list counter by 1
                b+=1
                    
            
        elif sp in (3,6,9,12,15,18):
                ax = fig.add_subplot(6,3,sp)
                ax.plot(women_degrees['Year'], women_degrees[other_cats[c]], c=cb_dark_blue, label='Women', linewidth=3)
                ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[c]], c=cb_orange, label='Men', linewidth=3)
                ax.set_title(other_cats[c])
                #increment list counter by 1
                c+=1
        
        #remove box outline of every chart
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        
        #remove xaxis from all charts by using labelbottom=False
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)

        if sp in (1,3):
            ax.text(2005, 87, 'Men')
            ax.text(2002, 8, 'Women')
        elif sp in (16,17,18):
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            #include xaxis label for bottom-most charts
            if sp in (16,18):
                ax.tick_params(labelbottom=True)
                
        elif sp ==14:
            #include xaxis label for bottom-most charts
            ax.tick_params(labelbottom=True)
plt.show()
```
