# Storytelling Data Visualization on Exchange Rates
In this project, we seek to visualize a [dataset](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) on Euro daily exchange rates between 1999 and 2021. (All values are pegged to the Euro currency)

```python
import pandas as pd
from IPython.display import display 
exchange_rates= pd.read_csv('euro-daily-hist_1999_2020.csv')
display(exchange_rates.head())
display(exchange_rates.tail())
exchange_rates.info()
```

We realise that there are a total of 5699 rows but for many columns, the values are null as reflected in the `df.info()` function above.
The data type for each column is also a mix of `string` and `float` values. Let us proceed to clean the dataset by modifying the data type and removing unwanted rows.

```python
#rename the columns for US dollar and Period\\Unit
exchange_rates.rename(columns={'[US dollar ]':'US_dollar', 'Period\\Unit:': 'Time'}, inplace=True)
exchange_rates['Time']= pd.to_datetime(exchange_rates['Time'])

euro_to_dollar= pd.DataFrame(exchange_rates[['Time','US_dollar']])

euro_to_dollar.sort_values('Time', inplace=True)
euro_to_dollar.reset_index(drop=True, inplace=True)

print(euro_to_dollar['US_dollar'].value_counts())
#there are 62 counts of '-' in the US_dollar column
#remove these rows
euro_to_dollar=euro_to_dollar.loc[euro_to_dollar['US_dollar']!='-']
euro_to_dollar.info()
```

```python
euro_to_dollar['US_dollar'].value_counts()
# convert column to float dtype
euro_to_dollar['US_dollar']=pd.to_numeric(euro_to_dollar['US_dollar'])
#verify that US_dollar column values have been converted to float
euro_to_dollar.info()
```

```python
#create a new column containing the 30 day average
euro_to_dollar['rolling']=euro_to_dollar['US_dollar'].rolling(30).mean()
```

Let us use the data on US_dollar to show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). For better representation, we will attempt to maximize the data-ink ratio, apply Gestalt principles and guide the audience through pre-attentive attributes. 

Plan:
Create 3 plots with each representing a specific administration and highlight the area of timeline it represents.

```python
#Bush Administration(Jan 2001- Jan 2009)
bush_start=euro_to_dollar['Time']>=pd.Timestamp(2001, 1, 20)
bush_end=euro_to_dollar['Time']<pd.Timestamp(2009, 1, 20)
bush_condition=bush_start&bush_end
bush=euro_to_dollar.loc[bush_condition]
bush.reset_index(drop=True, inplace=True)
display(bush.head())
print(bush['Time'].describe())

#Obama Administration(Jan 2009- Jan 2017)
obama_start=euro_to_dollar['Time']>=pd.Timestamp(2009, 1, 20)
obama_end=euro_to_dollar['Time']<pd.Timestamp(2017, 1, 20)
obama_condition=obama_start&obama_end
obama=euro_to_dollar.loc[obama_condition]
obama.reset_index(drop=True, inplace=True)
display(obama.head())
print(obama['Time'].describe())


#Trump Administration(Jan 2017- Jan 2021)
trump_start=euro_to_dollar['Time']>=pd.Timestamp(2017, 1, 20)
trump_end=euro_to_dollar['Time']<pd.Timestamp(2021, 1, 20)
trump_condition=trump_start&trump_end
trump=euro_to_dollar.loc[trump_condition]
trump.reset_index(drop=True, inplace=True)
display(trump.head())
print(trump['Time'].describe())
```

From the data above, it seems like we are missing some dates as the total count for bush administration is different from obama administration. Additionally, the data from 2021-01-09 to 2021-01-20 is missing. In order to account for this, we will also end the other administrations on the 8th of january for a better comparison.

```python
#Bush Administration(Jan 2001- Jan 2009)
bush_start=euro_to_dollar['Time']>=pd.Timestamp(2001, 1, 20)
bush_end=euro_to_dollar['Time']<pd.Timestamp(2009, 1, 8)
bush_condition=bush_start&bush_end
bush=euro_to_dollar.loc[bush_condition]
bush.reset_index(drop=True, inplace=True)
display(bush.head())
print(bush['Time'].describe())

#Obama Administration(Jan 2009- Jan 2017)
obama_start=euro_to_dollar['Time']>=pd.Timestamp(2009, 1, 20)
obama_end=euro_to_dollar['Time']<pd.Timestamp(2017, 1, 8)
obama_condition=obama_start&obama_end
obama=euro_to_dollar.loc[obama_condition]
obama.reset_index(drop=True, inplace=True)
display(obama.head())
print(obama['Time'].describe())


#Trump Administration(Jan 2017- Jan 2021)
trump_start=euro_to_dollar['Time']>=pd.Timestamp(2017, 1, 20)
trump_end=euro_to_dollar['Time']<pd.Timestamp(2021, 1, 8)
trump_condition=trump_start&trump_end
trump=euro_to_dollar.loc[trump_condition]
trump.reset_index(drop=True, inplace=True)
display(trump.head())
print(trump['Time'].describe())
```

```python
#Let us now begin plotting the graphs
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import matplotlib.style as style
import numpy as np
#allow jupyter to generate graphs
%matplotlib inline
style.use('fivethirtyeight')
fig, ax = plt.subplots(2,3, figsize=(18,8))
date_form = DateFormatter("%Y")
ax[0,0].text(x=pd.Timestamp("2001-01-20"), y=2.2, s='EURO-USD rate averaged 1.22 under the last three US presidents', 
             size=30, weight='bold')
ax[0,0].text(x=pd.Timestamp("2001-01-20"), y=2.1, s='EURO-USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017)', 
             size=24)
ax[0,0].text(x=pd.Timestamp("2001-01-20"), y=2, s='and Donald Trump (2018-2021)', 
             size=24)

#bush graph
ax[0,0].plot(bush['Time'], bush['rolling'], color='purple')
ax[0,0].text(x=pd.Timestamp("2004-03-20"), y=1.8, s='BUSH', size=20, weight='bold')
ax[0,0].locator_params(axis='y', nbins=4)
ax[0,0].xaxis.set_major_formatter(date_form)
# Ensure ticks fall once every other year (interval=2) 
ax[0,0].xaxis.set_major_locator(mdates.YearLocator(2))
ax[0,0].tick_params(axis='x', colors='grey')
ax[0,0].tick_params(axis='y', colors='grey')

#obama graph
ax[0,1].plot(obama['Time'], obama['rolling'], color='orange')
ax[0,1].set_ylim([0.8,1.6])
ax[0,1].text(x=pd.Timestamp("2012-03-20"), y=1.8, s='OBAMA', size=20, weight='bold')
ax[0,1].locator_params(axis='y', nbins=4)
ax[0,1].xaxis.set_major_formatter(date_form)
# Ensure ticks fall once every other year (interval=2) 
ax[0,1].xaxis.set_major_locator(mdates.YearLocator(2))
ax[0,1].tick_params(axis='x', colors='grey')
ax[0,1].tick_params(axis='y', colors='grey')

#trump graph
ax[0,2].plot(trump['Time'], trump['rolling'], color='blue')
ax[0,2].set_ylim([0.8,1.6])
ax[0,2].text(x=pd.Timestamp("2018-10-20"), y=1.8, s='TRUMP', size=20, weight='bold')
ax[0,2].locator_params(axis='y', nbins=4)
ax[0,2].xaxis.set_major_formatter(date_form)
# Ensure ticks fall once every other year (interval=2) 
ax[0,2].xaxis.set_major_locator(mdates.YearLocator(1))
ax[0,2].tick_params(axis='x', colors='grey')
ax[0,2].tick_params(axis='y', colors='grey')

#del 2 other axes
fig.delaxes(ax[1,0])
fig.delaxes(ax[1,2])

ax[1,1].set_position([0.08,0,0.88,.4])

#combining all three graphss
start=bush_start
end=trump_end
cond=start&end
combined=euro_to_dollar.loc[cond]
ax[1,1].plot(combined['Time'], combined['rolling'], color='orange', alpha=0)
ax[1,1].plot(bush['Time'], bush['rolling'], color='purple')
ax[1,1].plot(obama['Time'], obama['rolling'], color='orange')
ax[1,1].plot(trump['Time'], trump['rolling'], color='blue')
ax[1,1].locator_params(axis='y', nbins=4)
ax[1,1].tick_params(axis='y', colors='grey')
ax[1,1].xaxis.grid(False)
ax[1,1].set_xticks([])
plt.show()
```
